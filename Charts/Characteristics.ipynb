```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D   
```

```python
plt.xlim(-0.1, 1.1)
plt.ylim(-0.1, 1.1)

#boundary
xb = [0, 0, 1, 1, 0]
yb = [0, 1, 1, 0, 0]
plt.plot(xb, yb, color = 'gray')

# t grid
x_special1 = [0.2, 0.2]
y_special1 = [0, 1]
plt.plot(x_special1, y_special1, color = 'gray')

x_special2 = [0.8, 0.8]
y_special2 = [0, 1]
plt.plot(x_special2, y_special2, color = 'gray')

# t + h
x2 = [0, 0.8]
y2 = [0.2, 1]
plt.plot(x2, y2, color = '#4b0082', label = "x + h", linewidth=3)

# t - h
x1 = [0.2, 1]
y1 = [0 , 0.8]
plt. plot(x1, y1, color = 'red', label = "x - h", linewidth=3)

plt.legend(bbox_to_anchor=(0.95, 0.97),
           bbox_transform=plt.gcf().transFigure)

# boundary lines
x_s1 = [0.8, 1]
y_s1 = [1, 1]
plt.plot(x_s1, y_s1, color = 'black', linewidth=3)

#x_t1 = [1, 1] # "#" - because we need function s(t)
#y_t1 = [0.8, 1]
#plt.plot(x_t1, y_t1, color = 'black')

x_s0 = [0, 0.2]
y_s0 = [0, 0]
plt.plot(x_s0, y_s0, color = 'black', linewidth=3)

#x_t0 = [0, 0]
#y_t0 = [0, 0.2]
#plt.plot(x_t0, y_t0, color = 'black')

#boundary values
x_axis = [0, 0.2, 0.8, 1]
y_axis = [0, 0.2, 0.8, 1]
x_values = [0 , "h", "1 - h", 1]
y_values = ["0", "", "", 1]
plt.xticks(x_axis, x_values)
plt.yticks(y_axis, y_values)

plt.xlabel("x")
plt.ylabel("s")

plt.rcParams['figure.figsize'] = [10, 8]
plt.grid()
plt.show()
```
