```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.set_xlim([-6,6])
ax.set_ylim([-6,6])
ax.set_zlim([-6, 6])

x = np.linspace(-4, 4, 30)
z = np.linspace(-6.1, 7, 30)
X, Z = np.meshgrid(x, z)
Y = 0*X
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('Ox')
ax.set_ylabel('Oy')
ax.set_zlabel('Oz')


x1 = []
y1 = []
for i in range(len(z)):
    x1.append(0)
    y1.append(3.5)
ax.plot3D(x1, y1, z, color = 'black', linewidth=2)

x_axis = [-4, 0, 4]
x_values = ["-a", "0", "a"]
y_axis = [0, 3.5]
y_values = ["0", "b"]
z_axis = [0]
z_values = ["0"]
plt.xticks(x_axis, x_values)
plt.yticks(y_axis, y_values)
ax.set_zticks([0])
ax.set_zticklabels(["0"])

plt.show()

#run in spider and change angle
#change size
```
